### In this project i will use a KNeighborsClassifier on the iris dataset, to classify the three types of iris species based on thier petals and sepals length and width.
#### Load the iris dataset from sklearn and show its keys.

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.keys())
```

#### Import the pandas library to create  a dataframe from the iris dataset using the data and the feature names, then append our target feature to the dataframe, and show the first 5 rows.

```python
import pandas as pd

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df["species"] = iris.target
df.head()
```

#### Check shape of the dataset.

```python
df.shape
```

#### Show the iris dataset description, which show that there are three different types of iris species which are:
#### 1- Iris-Setosa.
#### 2- Iris-Versicolour.
#### 3- Iris-Virginica.
#### Each one of them have different length and width for its petal and sepal leaves.

```python
print(iris.DESCR)
```

#### Show Summary Statistics for our dataset.

```python
df.describe()
```

#### Plot our dataframe using a scatter_matrix to see the distribution, the scatter_matrix plot a histogram for each feature and a scatter plot for each feature against another for the 3 types of iris species.
#### To see the distribution for each of the plant types on the scatter plot, we create an object and map each plant type with a different color.
#### From looking at the scatter plots we can see that petal's length and width are highly correlated, you can easily separate each class from another, which mean these 2 features will be our X array ( the predictors ).

```python
import matplotlib.pyplot as plt
colors = df["species"].replace({0: "red", 1: "blue", 2: "green"})
pd.plotting.scatter_matrix(df, figsize=(15,15), c = colors)
plt.show()
```

#### Create the Two arrays the predictors and the target, X and y respectively.

```python
X = df[['petal length (cm)', 'petal width (cm)']].values
y = df['species'].values
```

#### Import the train_test_split function to split our 2 arrays into train and test sets.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1, stratify = y)
```

#### Import our classifier and train it on the train set then check its accuracy score.

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(X_train, y_train)
print("Accuracy score:", kn.score(X_test, y_test))
```

#### Let's see how well our model predictions did, by using a confusion matrix and plot it.
#### We can see that the model predicted 1 wrong.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
y_predict = kn.predict(X_test)
print(confusion_matrix(y_test, y_predict))

ConfusionMatrixDisplay.from_predictions(y_test, y_predict)
plt.show()
```

#### Now lets create a new model with n_neighbors = 3 and check its average score using  cross_val_score function with 5 kfold splits.
#### Note: our first model used n_neighbors = 5 which was the default attribute.

```python
from sklearn.model_selection import cross_val_score

kn_cv = KNeighborsClassifier(n_neighbors = 3)
cv_scores = cross_val_score(kn_cv, X, y, cv = 5)
print("Cross validation score:", cv_scores.mean())
```

#### Create a third model, but this time we split the data into train and test sets using kfold not train_test_split, and check its average score using cross_val_score function with kfolds = 5.

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 5, shuffle = True, random_state = 1)
for train_set, test_set in kf.split(X):
    X_train, X_test = X[train_set], X[test_set]
    y_train, y_test = y[train_set], y[test_set]
    
kn_kf = KNeighborsClassifier(n_neighbors = 3)
kn_kf.fit(X_train, y_train)
print("Accuracy score:", kn_kf.score(X_test, y_test))
cv_scores2 = cross_val_score(kn_kf, X, y, cv = 5)
print("Cross validation score:", cv_scores2.mean())
```

#### So how to determine which n_neighbors = ? will give the optimal result ?.
#### Sklearn have a GridSearchCV function to check which attributes are the best for our model's hyper parameters.
#### Import GridSearchCV and create a forth model and we create a param_grid to to find the optimal attribute for n_neighbors

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
kn2 = KNeighborsClassifier()

param_grid = {'n_neighbors': np.arange(2, 10)}

kn_gs = GridSearchCV(kn2, param_grid, cv = 5)
kn_gs.fit(X, y)

print("Best n_neighbors: ", kn_gs.best_params_)
```

#### GridSearchCV found best attribute is {'n_neighbors': 4}.
#### So we create our final model and use n_neighbors = 4.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1, stratify = y)
kn_final = KNeighborsClassifier(n_neighbors = 4)
kn_final.fit(X_train, y_train)

y_predict = kn_final.predict(X)
print("Accuracy score:", kn_final.score(X_test, y_test))
```

#### Create a new datapoint with petal's length and width and use our final model to predict its type and its prediction probability.

```python
new_data = np.array([[3.76, 1.20]])
print("Prediction", kn_final.predict(new_data))
print("prediction probability", kn_final.predict_proba(new_data))
```

#### We can also predict more than 1 datapoint at a time by using list of lists.

```python
new_data_1 = np.array([[3.76, 1.2], [5.25, 1.2], [1.58, 1.2]])
print("Prediction", kn_final.predict(new_data_1))
print("prediction probability\n", kn_final.predict_proba(new_data_1))
```
